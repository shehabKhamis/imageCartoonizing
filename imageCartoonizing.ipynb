**Loading Image**

```python
image=Image.open("/content/drive/MyDrive/CV.PNG")
print(image.format)
print(image.size)
print(image.mode)
image
```

**Converting image to grayscale**

```python
imageTemp=image.convert('L')
imageTemp
```

```python
data=asarray(imageTemp)
data.shape
```

**Applying median filter**

```python
medianFilter = cv2.medianBlur(data,5)
imgAfterMedian=Image.fromarray(medianFilter)
imgAfterMedian
```

**Edge detection using laplacian filter**

```python
laplacianFilter=cv2.Laplacian(medianFilter,-1)
imgAfterLaplacian=Image.fromarray(laplacianFilter)
imgAfterLaplacian
```

```python
(T, thresholding) = cv2.threshold(laplacianFilter, 7, 255,cv2.THRESH_BINARY)
imgAfterThresholding=Image.fromarray(thresholding)
imgAfterThresholding
```

```python
(Tinv, thresholdingInv) = cv2.threshold(laplacianFilter, 7, 255,cv2.THRESH_BINARY_INV)
imgAfterThresholdingInv=Image.fromarray(thresholdingInv)
imgAfterThresholdingInv
```

**Bilateral Filter**

```python
RGBimg=image.convert('RGB')
bilData=asarray(RGBimg)
painting=cv2.bilateralFilter(bilData,9,25,10)
imgAfterBil=Image.fromarray(painting)
imgAfterBil
```
